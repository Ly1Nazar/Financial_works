```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Тепер розглянемо ситуацію на українському ринку з прикладами</h2>


```python
years = np.arange(2010,2030,1)
```

```python
yearly_payment = 500*12
max_interest_insurance = 0.1
min_interest_insurance = 0.05
min_deposit_interest = 0.06
max_deposit_interest = 0.095
```

<h3>Підрахуемо прибуток депозиту на період 20 років</h3>

```python
min_result_deposit = 0
min_result_deposit_list = []
max_result_deposit = 0
max_result_deposit_list = []
for i in range(0,len(years)):
    min_result_deposit += yearly_payment*((1+min_deposit_interest)**i)
    min_result_deposit_list.append((yearly_payment*((1+min_deposit_interest)**i)))
    max_result_deposit += yearly_payment*((1+max_deposit_interest)**i)
    max_result_deposit_list.append((yearly_payment*((1+max_deposit_interest)**i)))
print(min_result_deposit_list)
```

<h3>Підрахуемо прибуток страхування на цей же період</h3>

```python
min_result_insurance = 0
min_result_insurance_list = []
max_result_insurance = 0
max_result_insurance_list = []
for c in range(0,len(years)):
    min_result_insurance += yearly_payment*((1+min_interest_insurance)**c)
    min_result_insurance_list.append((yearly_payment*((1+min_interest_insurance)**c)))
    max_result_insurance += yearly_payment*((1+max_interest_insurance)**c)
    max_result_insurance_list.append((yearly_payment*((1+max_interest_insurance)**c)))
```

```python
fig, ax = plt.subplots()
line1, = ax.plot(years, min_result_insurance_list, label='min insurance')
line1.set_dashes([2, 2, 10, 2])
#
line1, = ax.plot(years, min_result_deposit_list, label='min deposit')
line1.set_dashes([2, 2, 10, 2])
#
line1, = ax.plot(years, max_result_insurance_list, label='max insurance')
line1.set_dashes([2, 2, 10, 2])
#
line1, = ax.plot(years, max_result_deposit_list, label='max deposit')
line1.set_dashes([2, 2, 10, 2])

ax.legend()
plt.show()
```
